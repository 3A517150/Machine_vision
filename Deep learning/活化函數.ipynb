```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.array([-1,1,2.0])
y = step_function(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-5.0,5.0,0.1)
y = step_function(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-5.0,5.0,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
x = np.arange(-5,0.5,0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-0.1,5.0)
plt.show()
```
